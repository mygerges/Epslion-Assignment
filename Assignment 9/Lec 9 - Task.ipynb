```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
data = pd.read_csv('50ulke.csv')
data.head(10)
```

```python
data.info()
```

```python
sns.kdeplot(data['National Income'], shade=True, color='m');
```

```python
sns.distplot(data['National Income'], kde=False, color='m');
```

```python
sns.jointplot(x='Growth', y='Per person', data=data, kind='kde', color='m', height=8, ratio=6);
```

```python
sns.pairplot(data);
```

```python
sns.boxplot(x="Continent", y="Growth", data=data, palette='viridis');
```

```python
sns.violinplot(x="Continent", y="Growth", data=data, palette='viridis');
```

```python
sns.barplot(x="Continent", y="Growth", data=data, palette="viridis");
```

```python
sns.countplot(x="Continent", data=data, palette='viridis');
```

```python
sns.stripplot(x="Continent", y="Per person", data=data, palette="viridis");
```

```python
sns.swarmplot(x="Continent", y="Per person", data=data, palette="viridis");
```

```python
sns.lmplot(x="Per person", y="National Income", data=data, height=10, aspect=1.5);
```
